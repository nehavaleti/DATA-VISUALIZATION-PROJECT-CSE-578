```python
# remvoving missing-values with '?'
adult_data = adult_data[~adult_data.isin([' ?']).any(axis=1)]
```

```python
# dropping any other missing values
adult_data.dropna()
```

## DATA TRANSFORMATION


```python
# First step creating a separate dataset for the ratio attributes
new_dataset = adult_data[ [ 0, 2, 4, 10, 11, 12  ] ]
new_dataset
```

```python
categorical_attributes = [1, 5, 6, 7, 8, 9, 13]

for i in categorical_attributes:
    value_counts = adult_data[i].value_counts()  
    print( i,value_counts) 
```

```python
# Encoding categorical variables and 
#them combining them to the ratio_data(new dataset)

for i in categorical_attributes:
    encoding = pd.get_dummies((adult_data[i]))
    encoding = encoding.astype(int)
    new_dataset = pd.concat([new_dataset, encoding],axis=1)

new_dataset
```

```python
# Mapping Ordinal Values 
# Unique values in column 3 
adult_data[3].unique().tolist()
```

```python
edu_map = {
    ' Preschool':1,
    ' 1st-4th':2,
    ' 5th-6th':3,
    ' 7th-8th':4,
    ' 9th':5,
    ' 10tk':6,
    ' 11th':7,
    ' 12th':8,
    ' HS-grad':9,
    ' Some-College':10,
    ' Assoc-acdm': 11,
    ' Assoc-voc': 11,
    ' Bachelors': 12,
    ' Masters': 13,
    ' Prof-school':14,
    ' Doctorate': 15,
}
new_dataset.loc[:, 'Education'] = adult_data[3].map(edu_map)
```

```python
new_dataset
```

```python
# Mapping the column 14
adult_data[14].unique().tolist()
```

```python
income_map = {
    ' <=50K':0,
    ' >50K' :1 
}
new_dataset.loc[:, 'Income'] = adult_data[14].map(income_map)
new_dataset
```
